```python
import aiohttp
import asyncio
import pandas as pd

# Configurações da API IGDB
BASE_URL = "https://api.igdb.com/v4/games"

client_id = Credentials.twitch_igdb_id
access_token = Credentials.twitch_igdb_key

HEADERS = {
    "Client-ID": client_id,
    "Authorization": f"Bearer {access_token}",
}

async def fetch_game_data(session, nome_jogo, ano_jogo):
    """Faz uma requisição assíncrona para buscar informações do jogo na API IGDB."""
    body = f'search "{nome_jogo}"; where release_dates.y = {ano_jogo}; fields release_dates.y, franchises.slug, game_modes.slug, genres.slug, keywords.slug, name, slug, platforms.slug, player_perspectives.slug, total_rating, total_rating_count, summary, themes.slug, url, collections.slug;'

    try:
        async with session.post(BASE_URL, headers=HEADERS, data=body, timeout=30) as response:
            if response.status == 200:
                games_data = await response.json()
                print(f"Requisição ({response.status}) para {nome_jogo} ({ano_jogo})")
                if len(games_data)>=1:
                    return { "nome": nome_jogo, "ano": ano_jogo, "dados": games_data[0]}
                else:
                    print(f"Lista inexistente para {nome_jogo} ({ano_jogo})")
                    return { "nome": nome_jogo, "ano": ano_jogo, "dados": None }
            else:
                print(f"❌ Erro na requisição ({response.status}) para {nome_jogo} ({ano_jogo})")
                return { "nome": nome_jogo, "ano": ano_jogo, "dados": None }
    except Exception as e:
        print(f"⚠️ Erro inesperado para {nome_jogo} ({ano_jogo}): {e}")
        return { "nome": nome_jogo, "ano": ano_jogo, "dados": None }

async def process_games(df):
    """Executa as buscas em batchs de 7 jogos assíncronamente."""
    batch_size = 7
    results = []

    async with aiohttp.ClientSession() as session:
        for i in range(0, len(df), batch_size):
            batch = df.iloc[i:i + batch_size]
            tasks = [fetch_game_data(session, row["nome"], row["ano"]) for _, row in batch.iterrows()]
            batch_results = await asyncio.gather(*tasks)
            results.extend(batch_results)

    return results

# Função principal para rodar o processo
async def main():
    resultados = await process_games(df)

    for result in resultados:
        print(f"\n🎮 {result['nome']} ({result['ano']}):")
        print(result["dados"] if result["dados"] else "Nenhum dado encontrado.")

# Executando o código assíncrono
await main()
```

```python
import aiohttp
import asyncio
import pandas as pd

# Configurações da API IGDB
BASE_URL = "https://api.igdb.com/v4/games"

client_id = Credentials.twitch_igdb_id
access_token = Credentials.twitch_igdb_key

HEADERS = {
    "Client-ID": client_id,
    "Authorization": f"Bearer {access_token}",
}

async def fetch_game_data(session, nome_jogo, ano_jogo):
    """Faz uma requisição assíncrona para buscar informações do jogo na API IGDB."""
    body = f'search "{nome_jogo}"; where release_dates.y = {ano_jogo}; fields release_dates.y, franchises.slug, game_modes.slug, genres.slug, keywords.slug, name, slug, platforms.slug, player_perspectives.slug, total_rating, total_rating_count, summary, themes.slug, url, collections.slug;'

    try:
        async with session.post(BASE_URL, headers=HEADERS, data=body, timeout=30) as response:
            if response.status == 200:
                games_data = await response.json()
                print(f"✅ Sucesso ({response.status}) para {nome_jogo} ({ano_jogo})")

                if games_data:
                    return {
                        "nome": nome_jogo,
                        "ano": ano_jogo,
                        "dados": games_data[0]  # Pegamos apenas o primeiro resultado
                    }
                else:
                    print(f"⚠️ Nenhum resultado encontrado para {nome_jogo} ({ano_jogo})")
                    return { "nome": nome_jogo, "ano": ano_jogo, "dados": None }
            else:
                print(f"❌ Erro {response.status} para {nome_jogo} ({ano_jogo})")
                return { "nome": nome_jogo, "ano": ano_jogo, "dados": None }
    except Exception as e:
        print(f"⚠️ Erro inesperado para {nome_jogo} ({ano_jogo}): {e}")
        return { "nome": nome_jogo, "ano": ano_jogo, "dados": None }

async def process_games(df):
    """Executa as buscas em batchs de 7 jogos assíncronamente."""
    batch_size = 7
    results = []

    async with aiohttp.ClientSession() as session:
        for i in range(0, len(df), batch_size):
            batch = df.iloc[i:i + batch_size]
            tasks = [fetch_game_data(session, row["nome"], row["ano"]) for _, row in batch.iterrows()]
            batch_results = await asyncio.gather(*tasks)
            results.extend(batch_results)

    return results

async def main():
    resultados = await process_games(df)

    # Criando DataFrame com os dados filtrados
    dados_filtrados = [
        {
            "nome": r["nome"],
            "ano": r["ano"],
            "slug": r["dados"].get("slug", "N/A") if r["dados"] else "N/A",
            "rating": r["dados"].get("total_rating", "N/A") if r["dados"] else "N/A",
            "platforms": r["dados"].get("platforms", "N/A") if r["dados"] else "N/A",
            "summary": r["dados"].get("summary", "N/A") if r["dados"] else "N/A",
            "url": r["dados"].get("url", "N/A") if r["dados"] else "N/A",
        }
        for r in resultados if r["dados"]  # Apenas jogos encontrados
    ]

    df_resultados = pd.DataFrame(dados_filtrados)

    # Salvando os dados em CSV
    df_resultados.to_csv("games_data.csv", index=False, encoding="utf-8")
    print("\n📁 Dados salvos em 'games_data.csv'")

    # Exibindo DataFrame final
    print("\n📌 DataFrame Final:")
    print(df_resultados)

# Executando o código assíncrono
asyncio.run(main())
```
