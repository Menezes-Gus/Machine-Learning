```python
import kagglehub
from kagglehub import config
from kagglehub import KaggleDatasetAdapter
import pandas as pd
import numpy as np
import os
import sys
from ast import literal_eval
sys.path.append("..")
from credentials import Credentials

pd.set_option("display.max_columns",None)
```

```python
config.set_kaggle_credentials(Credentials.username, Credentials.key)
```

```python
df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "srgiomanhes/steam-games-dataset-2025",
    "steam_games.csv",
)
```

```python
df.head()
```

```python
for col in ['developers', 'publishers', 'categories', 'genres', 'platforms']:
    df[col] = df[col].apply(literal_eval)
```

```python
# Explodindo as colunas de listas para múltiplas linhas
df_explode = df.explode('developers')
df_explode = df_explode.explode('publishers')
df_explode = df_explode.explode('categories')
df_explode = df_explode.explode('genres')
df_explode = df_explode.explode('platforms')
```

```python
df_explode
```

```python
df_explode = df_explode[~df_explode['genres'].isin(['Early Access','Design & Illustration','Photo Editing','Utilities','Video Production','Game Development','Animation & Modeling','Audio Production','Software Training','Web Publishing','Education','Accounting','Movie','Documentary','Tutorial','360 Video'])]
```

```python

# Criando dummies (one-hot encoding)
df_final = pd.get_dummies(df_explode, columns=['developers'])
```

```python
df_final = pd.get_dummies(df_final, columns=['publishers'])
df_final = pd.get_dummies(df_final, columns=['categories'])
df_final = pd.get_dummies(df_final, columns=['genres'])
df_final = pd.get_dummies(df_final, columns=['platforms'])
```

```python
# Agrupando de volta por jogo, pegando o máximo (1 se pelo menos um jogo tem aquele atributo)
df_final = df_final.groupby('Jogo').max()

print(df_final)
```
