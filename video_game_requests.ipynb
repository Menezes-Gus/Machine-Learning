```python
import requests
import json

# URL da API
url = "https://jrpgc.com/jrpg-database/?_release_date=1984.00,2025.00"

# Cabeçalhos (simulando um navegador real)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36",
    "Content-Type": "application/json"
}

# Corpo da requisição (ajuste conforme necessário)
payload = {
    "action": "facetwp_refresh",
    "data": {},
    "extras": {},
    "sort": "default",
    "facets": {},
    "categories": [],
    "collections": [],
    "developers": [],
    "game_modes": [],
    "genres": [],
    "pagination": [],
    "per_page": [],
    "platform": [],
    "release_date": ["1984.00", "2025.00"],
    "results_count": [],
    "search": "",
    "sort_fp": [],
    "tags": "",
    "first_load": 0,
    "frozen_facets": {},
    "_release_date": "1984.00,2025.00",
    "uri": "jrpg-database",
    "paged": 1,  # Inicialmente, página 1
    "soft_refresh": 1,
    "template": "wp",
}

# Lista para armazenar todos os jogos
todos_os_jogos = []

# Loop para baixar todas as páginas
while True:
    response = requests.post(url, headers=headers, json=payload)
    
    # Verificando o código de status e o conteúdo da resposta
    if response.status_code != 200:
        print(f"Erro ao buscar os dados: {response.status_code}")
        print(f"Conteúdo da resposta: {response.text}")
        break
    
    # Tentar converter a resposta em JSON
    try:
        data = response.json()  # Converte o JSON para dicionário Python
    except json.JSONDecodeError as e:
        print(f"Erro ao decodificar JSON: {e}")
        print(f"Conteúdo da resposta: {response.text}")
        break
    
    # Verifique se há jogos na resposta
    jogos = data.get("results", [])  # Pode precisar ajustar conforme a estrutura
    if not jogos:
        break  # Sai do loop quando não houver mais jogos
    
    todos_os_jogos.extend(jogos)  # Adiciona os jogos à lista final
    print(f"Baixados {len(todos_os_jogos)} jogos...")
    
    # Avança para a próxima página
    payload["paged"] += 1  # Incrementa a página

# Exibir alguns dos jogos baixados
print(json.dumps(todos_os_jogos[:5], indent=4, ensure_ascii=False))
```
